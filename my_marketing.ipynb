Manual Segment:

1) create luxury and non luxury brand out of 'make' column.

2) create segment for hoursepower of car

```python
# List of luxury car makes
luxury_brands = ['mercedes-benz', 'bmw', 'audi', 'porshe', 'jaguar']

# Create a new column 'luxury' based on the 'make' column
data['luxury'] = data['make'].apply(lambda x: 'Luxury' if x.lower() in luxury_brands else 'Non-Luxury')

# Display the result
print(data[['make', 'luxury']])
```

```python
# Define criteria for segmentation based on 'horsepower'
low_hp_threshold = 100
high_hp_threshold = 200

# Create a new column 'horsepower_segment' based on 'horsepower'
data['horsepower_segment'] = pd.cut(data['horsepower'],
                                   bins=[-float('inf'), low_hp_threshold, high_hp_threshold, float('inf')],
                                   labels=['Low Horsepower', 'Medium Horsepower', 'High Horsepower'])

# Display the result
print(data[['make', 'horsepower', 'horsepower_segment']])
```

```python
# Define criteria for segmentation based on 'horsepower'
low_hp_threshold = 100
high_hp_threshold = 200

# Create a new column 'horsepower_segment' based on 'horsepower'
data['horsepower_segment'] = pd.cut(data['horsepower'],
                                   bins=[-float('inf'), low_hp_threshold, high_hp_threshold, float('inf')],
                                   labels=['Low Horsepower', 'Medium Horsepower', 'High Horsepower'])

# Create a bar plot for the segments based on 'horsepower' and 'make'
plt.figure(figsize=(15, 8))
sns.countplot(x='make', hue='horsepower_segment', data=data, palette='viridis')
plt.title('Car Segmentation by Make and Horsepower')
plt.xlabel('Car Make')
plt.ylabel('Count')
plt.legend(title='Horsepower Segment')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
